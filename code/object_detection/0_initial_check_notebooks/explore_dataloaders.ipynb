# Dataloader DFire

```python
loader = modules.dataloaders.get_val_loader()
```

### Plot Images with Bounding Boxes

```python
for batch_idx, (img, label) in enumerate(loader):
       
    if batch_idx == 48:
        print(f'Batch size equal to img.shape[0] = {img.shape[0]}')
        print(f'Batch images shape = {img.shape}')
        plt.subplots(2, 4, figsize=(10,6))
        for i in range(8):
            pic = utils.plot_dataset_img(img[i], label[i], grid=True)
            plt.subplot(2, 4, i+1)
            plt.imshow(pic)
        plt.tight_layout()
        plt.show()
        break
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = '../../datasets/dfire_mini/4ex/images/WEB10100.jpg'
img_np = cv2.imread(img)
img_np = cv2.cvtColor(img_np, cv2.COLOR_BGR2RGB)    
```

# Scale [0, 255]

Floor division to get int in range [0, 15]
Scale back to [0, 255] multiplying by 16, althoug new range will be [0, 240]

```python
img_np_4bit = (img_np // 16).astype(np.uint8)
img_np_4bit = img_np_4bit * 16
```

```python
plt.subplots(1, 2, figsize=(8,6))

plt.subplot(1, 2, 1)
plt.title("8 Bit")
plt.imshow(img_np)

plt.subplot(1, 2, 2)
plt.title("4 Bit")
plt.imshow(img_np_4bit)

plt.tight_layout()
plt.show()
```

# Check the same in range [0, 1]

```python
img_np_4bit = ((img_np // 16).astype(np.uint8) / 15)
img_np = img_np / 255    
```

```python
plt.subplots(1, 2, figsize=(8,6))

plt.subplot(1, 2, 1)
plt.title("8 Bit")
plt.imshow(img_np)

plt.subplot(1, 2, 2)
plt.title("4 Bit")
plt.imshow(img_np_4bit)

plt.tight_layout()
plt.show()
```
